```python
import tkinter as tk
from tkinter import ttk
from tkinter import * 
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

window = tk.Tk()


window.geometry("500x350") 
window.title("Fizyka - kinematyka")

frame = Frame(window)
frame.pack()
  
bottomframe = Frame(window)
bottomframe.pack(expand=False, fill="x")


tekst_wstepny = Label(frame, text = " \nProgram przedstawiający podstawy kinematyki \n")
tekst_wstepny.pack(side=TOP)

tabControl = ttk.Notebook(bottomframe)
  
tab1 = ttk.Frame(tabControl)
tab2 = ttk.Frame(tabControl)
  
tabControl.add(tab1, text ='Ruch jednostajny prostoliniowy')
tabControl.add(tab2, text ='Ruch jednostajnie przyspieszony')
tabControl.pack(expand=False , fill="x")


def zad1():
     
    # convert kg to gram
    predkosc = float(e2_value.get())
     
    # convert kg to pound
    czas = float(e4_value.get())
     
    # convert kg to ounce
    droga = predkosc * czas
     
    # Enters the converted street to
    # the text widget
    zad1_wynik.delete("1.0", END)
    zad1_wynik.insert(END,droga)


    t=int(e4_value.get())
    t=t+1
    V=predkosc
    czas = []
    droga = []

    for x in range(t):
        tw=x
        czas.append(x)
        droga.append(V*tw)

    data = []
    data.append(czas)
    data.append(droga)
    df = pd.DataFrame (data).transpose()
    df.columns=['czas', 'droga']

    class Table:

        def __init__(self,root):

            # code for creating table
            for i in range(total_rows):
                for j in range(total_columns):

                    if (i==0 and j==0):
                        self.e = Entry(root)
                        self.e.grid(row=i+1, column=j)
                        self.e.insert(END, "czas t[s]")
                    elif(i==1 and j==0):
                        self.e = Entry(root)
                        self.e.grid(row=i+1, column=j)
                        self.e.insert(END, "droga S[m]")
                    else:    
                        self.e = Entry(root)
                        self.e.grid(row=i+1, column=j)
                        self.e.insert(END, data[i][j])


    # find total number of rows and
    # columns in list
    total_rows = len(data)
    total_columns = len(data[0])

    # create root window
    root = tk.Tk()

    label_zad1_tabela = Label(root, text = "Tabela zależności drogi od czasu S(t):")
    label_zad1_tabela.grid(row = 0, column = 0, columnspan=total_columns, sticky = W)
    t = Table(root)
    
    label_zad1_tabela = Label(root, text = "\n")
    label_zad1_tabela.grid()
    
    wykres1 = plt.Figure(figsize=(5,4), dpi=100)
    ax3 = wykres1.add_subplot(111)
    ax3.scatter(df['czas'],df['droga'], color = 'r')
    ax3.set_xlabel('Czas [s]')
    ax3.set_ylabel('Droga [m]')
    ax3.set_title('Zależność drogi od czasu S(t)')

    scatter3 = FigureCanvasTkAgg(wykres1, root) 
    scatter3.get_tk_widget().grid(columnspan=total_columns)



# treść zadania
text_zad1_tresc = Text(tab1, height = "5", width = "60" )
tresc = """Oblicz jaką drogę przebędzie kulka poruszająca się ruchem jednostajnym prostoliniowym V po zadanym czasie t. 
Przedstaw w postaci tabeli drogę którą przebyła kulka po kolejnych sekundach ruchu oraz narysuj wykres zależności drogi od czasu S(t)."""
text_zad1_tresc.insert(tk.END, tresc)
# dane
label_zad1_dane = Label(tab1, text = "Podaj dane:")
label_zad1_predkosc = Label(tab1, text = "Prędkość V[m/s]=")
e2_value = StringVar()
entry_zad1_predkosc = Entry(tab1, textvariable = e2_value)

label_zad1_czas = Label(tab1, text = "Czas t[s]=")
e4_value = StringVar()
entry_zad1_czas = Entry(tab1, textvariable = e4_value)

# Create the Button Widget
button_zad1_oblicz = Button(tab1, text = "Oblicz",command = zad1)

# wynik
label_zad1_wynik = Label(tab1, text = "Wynik obliczeń:")

label_zad1_droga = Label(tab1, text = 'Droga S[m]=')
zad1_wynik = Text(tab1, height = 1, width = 15)
 
# grid method
text_zad1_tresc.grid(row = 0, column = 0, columnspan=5, padx = 5, pady = 5)
label_zad1_dane.grid(row = 1, column = 0, padx = 5, pady = 5, sticky = W)
label_zad1_predkosc.grid(row = 2, column = 0, padx = 5, pady = 5, sticky = W)
entry_zad1_predkosc.grid(row = 2, column = 1, padx = 5, pady = 5, sticky = W)

label_zad1_czas.grid(row = 3, column = 0, padx = 5, pady = 5, sticky = W)
entry_zad1_czas.grid(row = 3, column = 1, padx = 5, pady = 5, sticky = W)

button_zad1_oblicz.grid(row = 4, column = 1, padx = 5, pady = 5, sticky = W)

label_zad1_wynik.grid(row = 1, column = 3, padx = 5, pady = 5, sticky = W)
label_zad1_droga.grid(row = 2, column = 3, padx = 5, pady = 5, sticky = W)
zad1_wynik.grid(row = 2, column = 4)


def zad2():
     
    # convert kg to gram
    przyspieszenie = float(e22_value.get())
     
    # convert kg to pound
    czas = float(e42_value.get())
     
    # convert kg to ounce
    droga = (przyspieszenie * czas * czas)/2
     
    # convert kg to ounce
    predkosc = przyspieszenie * czas   
        
    # Result
    zad2_wynik.delete("1.0", END)
    zad2_wynik.insert(END,droga)
    
    zad2_wynik_predkosc.delete("1.0", END)
    zad2_wynik_predkosc.insert(END,predkosc)

    t=int(e42_value.get())
    t=t+1
    a=przyspieszenie
    czas = []
    droga = []
    predkosc = []
    
    for x in range(t):
        tw=x
        czas.append(x)
        droga.append((a*tw*tw)/2)
        predkosc.append(a*tw)
        
    data = []
    data.append(czas)
    data.append(droga)
    df = pd.DataFrame (data).transpose()
    df.columns=['czas', 'droga']

    data2 = []
    data2.append(czas)
    data2.append(predkosc)
    df2 = pd.DataFrame (data2).transpose()
    df2.columns=['czas', 'predkosc']
    
    class Table:

        def __init__(self,root,x,data,label1,label2):

            # code for creating table
            for i in range(total_rows):
                for j in range(total_columns):
                    if (i==0 and j==0):
                        self.e = Entry(root)
                        self.e.grid(row=i+1, column=total_columns*x+j)
                        self.e.insert(END, label1)
                    elif(i==1 and j==0):
                        self.e = Entry(root)
                        self.e.grid(row=i+1, column=total_columns*x+j)
                        self.e.insert(END, label2)
                    else:    
                        self.e = Entry(root)
                        self.e.grid(row=i+1, column=total_columns*x+j)
                        self.e.insert(END, data[i][j])


    # find total number of rows and
    # columns in list
    total_rows = len(data)
    total_columns = len(data[0])

    # create root window
    root = tk.Tk()

    label_zad2_tabela = Label(root, text = "Tabela zależności drogi od czasu S(t):")
    label_zad2_tabela.grid(row = 0, column = 0, columnspan=total_columns, sticky = W)
    t = Table(root,0,data,"czas t[s]","droga S[m]")
    
    label_zad2_tabela = Label(root, text = "\n")
    label_zad2_tabela.grid()
    
    zad2_wykres1 = plt.Figure(figsize=(5,4), dpi=100)
    ax2 = zad2_wykres1.add_subplot(111)
    ax2.scatter(df['czas'],df['droga'], color = 'r')
    ax2.set_xlabel('Czas [s]')
    ax2.set_ylabel('Droga [m]')
    ax2.set_title('Zależność drogi od czasu S(t)')

    scatter2 = FigureCanvasTkAgg(zad2_wykres1, root) 
    scatter2.get_tk_widget().grid(columnspan=total_columns)

    label_zad2_tabela2 = Label(root, text = "Tabela zależności prędkości od czasu V(t):")
    label_zad2_tabela2.grid(row = 0, column = total_columns+1, columnspan=total_columns, sticky = W)
    t2 = Table(root,1,data2,"||  czas t[s]","||  prędkość V[m/s]")
    
    label_zad2_tabela = Label(root, text = "\n")
    label_zad2_tabela.grid()
    
    zad2_wykres2 = plt.Figure(figsize=(5,4), dpi=100)
    ax3 = zad2_wykres2.add_subplot(111)
    ax3.scatter(df2['czas'],df2['predkosc'], color = 'g')
    ax3.set_xlabel('Czas [s]')
    ax3.set_ylabel('Prędkość [m/s]')
    ax3.set_title('Zależność prędkości od czasu V(t)')

    scatter3 = FigureCanvasTkAgg(zad2_wykres2, root) 
    scatter3.get_tk_widget().grid(row = 4, column = total_columns, columnspan=total_columns)

# treść zadania
text_zad2_tresc = Text(tab2, height = "7", width = "60" )
tresc2 = """Oblicz jaką drogę przebędzie kulka poruszająca się ruchem prostoliniowym jednostajnie przyspieszonym "a" po zadanym czasie t. Oblicz też jaką prędkość V osiągnie kulka po czasie t. Przedstaw w postaci tabeli drogę którą przebyła kulka po kolejnych sekundach ruchu oraz narysuj wykres zależności drogi od czasu S(t).Podobnie przedstaw w tabeli i wykresie zależność prędkości od czasu a(t)."""
text_zad2_tresc.insert(tk.END, tresc2)
# dane
label_zad2_dane = Label(tab2, text = "Podaj dane:")
label_zad2_przyspieszenie = Label(tab2, text = "Przyspieszenie a[m/s2]=")
e22_value = StringVar()
entry_zad2_przyspieszenie = Entry(tab2, textvariable = e22_value)

label_zad2_czas = Label(tab2, text = "Czas t[s]=")
e42_value = StringVar()
entry_zad2_czas = Entry(tab2, textvariable = e42_value)

# Create the Button Widget
button_zad2_oblicz = Button(tab2, text = "Oblicz",command = zad2)

# wynik
label_zad2_wynik = Label(tab2, text = "Wynik obliczeń:")

label_zad2_droga = Label(tab2, text = 'Droga S[m]=')
zad2_wynik = Text(tab2, height = 1, width = 15)

label_zad2_predkosc = Label(tab2, text = 'Prędkość V[m/s]=')
zad2_wynik_predkosc = Text(tab2, height = 1, width = 15)

# grid method
text_zad2_tresc.grid(row = 0, column = 0, columnspan=5, padx = 5, pady = 5)
label_zad2_dane.grid(row = 1, column = 0, padx = 5, pady = 5, sticky = W)
label_zad2_przyspieszenie.grid(row = 2, column = 0, padx = 5, pady = 5, sticky = W)
entry_zad2_przyspieszenie.grid(row = 2, column = 1, padx = 5, pady = 5, sticky = W)

label_zad2_czas.grid(row = 3, column = 0, padx = 5, pady = 5, sticky = W)
entry_zad2_czas.grid(row = 3, column = 1, padx = 5, pady = 5, sticky = W)

button_zad2_oblicz.grid(row = 4, column = 1, padx = 5, pady = 5, sticky = W)

label_zad2_wynik.grid(row = 1, column = 3, padx = 5, pady = 5, sticky = W)
label_zad2_droga.grid(row = 2, column = 3, padx = 5, pady = 5, sticky = W)
zad2_wynik.grid(row = 2, column = 4)
label_zad2_predkosc.grid(row = 3, column = 3, padx = 5, pady = 5, sticky = W)
zad2_wynik_predkosc.grid(row = 3, column = 4)


# Start the GUI
window.mainloop()
```
