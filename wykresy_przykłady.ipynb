```python
%matplotlib notebook

import pandas as pd
import matplotlib.pyplot as plt
from itertools import count
import random
from matplotlib.animation import FuncAnimation

x_vals = []
y_vals = []

y1=[1,4,5,2,1,7]

index = count()

def animate(i):
    
    x_vals.append(next(index))
    y_vals.append(y1[i]) 

    
    plt.cla()
    plt.plot(x_vals, y_vals)
    
ani = FuncAnimation(plt.gcf() , animate , interval=1000 )   
    
plt.plot(x_vals, y_vals)

plt.tight_layout()
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation, writers
import numpy as np
  
fig = plt.figure(figsize = (7,5))
axes = fig.add_subplot(1,1,1)
axes.set_ylim(0, 10)
palette = ['yellow', 'black', 'green', 
           'red', 'maroon', 'blue']
  
y1, y2, y3, y4, y5, y6 = [], [], [], [], [], []

y11, y21, y31, y41, y51, y61 = [1,2,3,5,7], [1.1,2,3.4,6,8], [2,3,6,8,9], [1,2,5,8,9], [1,2,3,7,6], [1,2,3,4,1]

def animation_function(i):
    y1 = y11[i]
    y2 = y21[i]
    y3 = y31[i]
    y4 = y41[i]
    y5 = y51[i]
    y6 = y61[i]
    
    plt.cla()    
    plt.xlabel("testowe kategorie - zmiana w czasie")
    plt.ylabel("testowe dane")
      
    plt.bar(["Dana1", "Dana2", "Dana3", 
             "Dana4", "Dana5", "Dana6"],
            [y1, y2, y3, y4, y5, y6],
            color = palette)
  
plt.title("Bar Chart Animation")
  
animation = FuncAnimation(fig, animation_function, 
                          interval = 1000)
plt.show()
```

```python
%matplotlib notebook

from wordcloud import WordCloud, STOPWORDS

import matplotlib.pyplot as plt

text="Woda, Powietrze, Ziemia, OgieÅ„, Trawa, Niebo, LÃ³d"
stopwords = set(STOPWORDS)
cloud = WordCloud(background_color="white", max_words=200, mask=None, 
        stopwords=stopwords, min_font_size=6, width=800, height=400)
 
wordcloud = cloud.generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
